```python
import pandas as pd
```

```python
spotify_data_track = pd.read_csv('track_data_final.csv')
print(spotify_data_track.head(10))
```

### Preguntas de practica (nivel facil)
1. ¿Cuál es el nombre del artista con mayor cantidad de seguidores en el archivo?
2. ¿Cuántas canciones explícitas (explicit=True) hay en el archivo?

```python
# 2 - cual es el nombre del artista con mayor cantidad de seguidores en el archivo?

artist_max_followers = (
    spotify_data_track.groupby('artist_name')['artist_followers']
    .max()
    .sort_values(ascending=False)
    .reset_index()
)

artist_max_followers.columns = ['Artista', 'Seguidores']
artist_max_followers['Seguidores'] = artist_max_followers['Seguidores'].apply(lambda x: f"{int(x):,}".replace(',', '.'))
print(artist_max_followers.head(5))
```

```python
# 2 - cuantas canciones explicitas (explicit= true) hay en el archivo
explicit_songs_count = spotify_data_track[spotify_data_track['explicit'] == True].shape[0]
print(f"Cantidad de canciones explícitas: {explicit_songs_count}")
```

### Pregunata de practica (nivel intermedio)

3. ¿Cuál es el género musical con mayor popularidad promedio entre las canciones del archivo?

4. ¿Qué tipo de álbum (album_type) tiene la mayor cantidad de canciones explícitas?

```python
# 3 - cual es el género musical con mayor pupularidad promedio entre las canciones del archivo ?

genres_max_popularity = (
    spotify_data_track.groupby('artist_genres')['track_popularity']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
    )

print(genres_max_popularity)
```

```python
# 4 - que tipo de álbum(album_type) tiene la mayor cantidad de canciones explícictas?

album_type_explicit_count = (
    spotify_data_track[spotify_data_track.explicit == True]
    .groupby('album_type')['explicit']
    .count()
    .sort_values(ascending=False)
    .reset_index()
    .head(5)
)

album_type_explicit_count.columns = ['tipo de album', 'cantidad de canciones explicitas']
album_type_explicit_count['cantidad de canciones explicitas'] = (
    album_type_explicit_count['cantidad de canciones explicitas']
    .apply(lambda x:f"{int(x):,}".replace(',','.'))
)
print(album_type_explicit_count)
```
